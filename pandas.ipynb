```python
#Importando o Pandas e o arquivo CSV:
import pandas as pd

# Importando o arquivo CSV
df = pd.read_excel('C:/Users/itarg/Downloads/AULA - PROGRAMACAO PARA BIG DATA/dados_vendas.xlsx')
```

```python
#Visualizando os primeiros registros do DataFrame:

print("Primeiras linhas do DataFrame:")
print(df.head())
```

```python
#Realizando algumas operações básicas:
# Verificando o número de linhas e colunas do DataFrame
print("\nNúmero de linhas e colunas:")
print(df.shape)

# Verificando os tipos de dados das colunas
print("\nTipos de dados das colunas:")
print(df.dtypes)

# Calculando estatísticas descritivas para as colunas numéricas
print("\nEstatísticas descritivas:")
print(df.describe())
```

```python
#Visualizando os dados em gráficos:
import matplotlib.pyplot as plt

# Criando um gráfico de barras para visualizar as vendas por produto

df.groupby('Produto')['Vendas'].sum().plot(kind='bar', color='skyblue')
plt.title('Vendas por Produto')
plt.xlabel('Produto')
plt.ylabel('Total de Vendas')
plt.show()

# Criando um gráfico de pizza para visualizar a distribuição de vendas por região
df.groupby('Região')['Vendas'].sum().plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen', 'lightblue', 'lightcoral'])
plt.title('Distribuição de Vendas por Região')
plt.ylabel('')
plt.show()
```
